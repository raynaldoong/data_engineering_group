# 2. Road Accident Dataset

## 2.1 Opening Dataset

```python
# Define the file path
filename = "raw_data/road_accident_2022.csv"

# Read the CSV file into a pandas DataFrame
df1 = pd.read_csv(filename)

# Display the DataFrame
df1
```

## 2.2 Dataset Overview

```python
df1.info()
```

```python
print(df1.dtypes)
```

## 2.3 Data Cleaning

### 2.3.1 Checking Duplicate Values

```python
from pyspark.sql.functions import count, col

# Convert Pandas DataFrame to Spark DataFrame
spark_df = spark.createDataFrame(df1)

# Iterate over each column and check for duplicates
for column_name in spark_df.columns:
    duplicates = spark_df.groupBy(column_name).agg(count("*").alias("count")).filter(col("count") > 1)
    num_duplicates = duplicates.count()
    print("Column '{}' has {} duplicates.".format(column_name, num_duplicates))
```

### 2.3.2 Checking Percentage of Null Values

```python
# Calculate the percentage of non-null values in each column
non_null_pct = df1.notnull().mean() * 100

# Print the percentage of non-null values in each column
print(non_null_pct)
```

### 2.3.3 Drop rows with missing OSGR Coordinates

```python
# Count the number of rows in the original DataFrame
num_rows_before = len(df1)

# Drop rows with missing location_easting_osgr and location_northing_osgr values
df1 = df1.dropna(subset=["location_easting_osgr", "location_northing_osgr"])

# Count the number of rows in the modified DataFrame
num_rows_after = len(df1)

# Count the number of rows that were dropped
num_rows_dropped = num_rows_before - num_rows_after

# Display the number of rows that were dropped
print(f"Dropped {num_rows_dropped} rows with missing values.")
```

### 2.3.4 Drop redundant columns

```python
df1 = df1.drop(["accident_year","day_of_week","status"], axis=1)
```

## 2.4 Data Transformation

### 2.4.1 Changing OSGR coordinates to longitude and lattitude

```python
# Define the OSGB36 and WGS84 coordinate systems
osgb36 = pyproj.Proj(init='epsg:27700')  # OSGB36 British National Grid
wgs84 = pyproj.Proj(init='epsg:4326')   # WGS84 global coordinate system

# Convert the OSGB36 coordinates to latitude and longitude
longitude, latitude = pyproj.transform(osgb36, wgs84, df1['location_easting_osgr'].values, df1['location_northing_osgr'].values)

# Add the latitude and longitude columns to the DataFrame
df1['latitude'] = latitude
df1['longitude'] = longitude

# Print the first few rows of the DataFrame with the new latitude and longitude columns
print(df1[['location_easting_osgr', 'location_northing_osgr', 'latitude', 'longitude']].head())
```

```python
df1 = df1.drop(["location_easting_osgr","location_northing_osgr"], axis=1)
```

### 2.4.2 Change date column to datetime

```python
df1
```

```python
df1['date'] = pd.to_datetime(df1['date'], dayfirst=True)
df1['time'] = pd.to_datetime(df1['time'], format='%H:%M').dt.time
```

## 2.5 Data Encoding

### 2.5.1 Open Accident Encoding

```python
# Read the Excel file into a pandas DataFrame
accident_encoding = pd.read_excel('raw_data/accident_encoding.xlsx')

# Display the DataFrame
accident_encoding
```

```python
# Get unique field names from accident_encoding DataFrame
field_names = accident_encoding['field name'].unique()

# Loop through each field name
for field_name in field_names:
    # Check if the field_name exists in df1
    if field_name in df1.columns:
        # Filter the accident_encoding DataFrame for the current field_name
        field_encoding = accident_encoding[accident_encoding['field name'] == field_name]

        # Create a dictionary for encoding, mapping code/format to label
        encoding_dict = field_encoding.set_index('code/format')['label'].to_dict()

        # Replace values in the df1 DataFrame column using the encoding_dict
        df1[field_name] = df1[field_name].map(encoding_dict).fillna(df1[field_name])
```

```python
# Display the updated df1 DataFrame
df1
```

# 3. Accident Vehicle Dataset

## 3.1 Opening dataset

```python
# Define the file path
filename = "raw_data/accident_vehicle_2022.csv"

# Read the CSV file into a pandas DataFrame
df2 = pd.read_csv(filename)

# Display the DataFrame
df2
```

## 3.2 Dataset overview

```python
df2.info()
```

```python
print(df2.dtypes)
```

# 3.3 Data Cleaning

### 3.3.1 Check for duplicate values

```python
# Iterate over each column and check for duplicates
for col in df2.columns:
    duplicates = df2[col].duplicated()
    num_duplicates = duplicates.sum()
    print("Column '{}' has {} duplicates.".format(col, num_duplicates))
```

### 3.3.2 Check for Percentage of Null values

```python
# Calculate the percentage of non-null values in each column
non_null_pct = df2.notnull().mean() * 100

# Print the percentage of non-null values in each column
print(non_null_pct)
```

### 3.3.3 Drop redundant columns

```python
df2 = df2.drop(["accident_year","accident_reference","vehicle_reference","status"], axis=1)
```

## 3.4 Data Encoding

### 3.4.1 Vehicle Encoding

```python
# Read the Excel file into a pandas DataFrame
vehicle_encoding = pd.read_excel('raw_data/vehicle_encoding.xlsx')

# Display the DataFrame
vehicle_encoding
```

### 3.4.2 Decode Columns

```python
# Step 1: Create a dictionary of dictionaries to store the encoding mappings
encoding_mappings = {}

# Step 2: Iterate through the vehicle_encoding DataFrame and populate the dictionary
for index, row in vehicle_encoding.iterrows():
    column_name = row['field name']
    code = row['code/format']
    label = row['label']
    
    if column_name not in encoding_mappings:
        encoding_mappings[column_name] = {}
    
    encoding_mappings[column_name][code] = label

# Step 3: Create a function to apply the decoding to the relevant columns in df2
def decode_column(value, encoding_map):
    if value in encoding_map:
        return encoding_map[value]
    return value

# Step 4: Apply the decoding function to the appropriate columns in df2
for column in encoding_mappings.keys():
    if column in df2.columns:
        df2[column] = df2[column].apply(lambda x: decode_column(x, encoding_mappings[column]))

# df2 will now have the decoded values from vehicle_encoding
```

```python
# Display the updated df2 DataFrame
df2
```

# 3.5 Data Aggregation

```python
#aggregate to columnar-style table
aggregated_df2 = df2.groupby('accident_index').agg(lambda x: list(x)).reset_index()
```
